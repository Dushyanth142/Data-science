```python
num_cols=df.select_dtypes(include=['int64','float64']).columns
num_cols
```

```python
for column in num_cols:
    df[column].plot(kind='hist',title=f'{column} Distribution',bins=20)
    plt.show()
```

```python
for column in num_cols:
    df[column].plot(kind='box',title=f'{column} Boxplot')
    plt.show()
```

```python
Bar chart analysis for categorical columns
```

```python
# identifying categorical columns
cat_cols=df.select_dtypes(include=['object']).columns
cat_cols
```

```python
for column in cat_cols:
    df[column].value_counts().plot(kind='bar',title=f'{column}')
    plt.show()
```

Standardization of Numerical Variables


```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
stdized_df=df.copy()
stdized_df[num_cols]=Scaler.fit_transform(df[num_cols])

# Before and After Comparisions
comparision=pd.DataFrame({
    'Original': df[num_cols[0]],
    'Standardized':stdized_df[num_cols[0]]
})
comparision
```

```python
comparision.head()
```

Conversion of categorical data into dummy variables


```python
# using one-hot encoding on categorical columns
df_encode=pd.get_dummies(df,columns=cat_cols,drop_first=True)
df_encode
```

```python
df_encode.head()
```

Conclusions


1.Histograms and boxplots are useful for visualizing data distribution and spotting any skewness or outliers.
2.One-hot encoding transforms categorical variables into numerical format that is suitable for machine learning models.
3.Standardization ensures numerical features have a mean of 0 and standard deviation of 1,which is important for many ML algorithms.
4.Descriptive Statitics offer insights into the central tendency and spread of numerical data.
