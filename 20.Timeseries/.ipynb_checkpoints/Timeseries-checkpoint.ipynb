```python
# Assuming 'currency' DataFrame is loaded correctly
currency = pd.read_csv('exchange_rate.csv', parse_dates=True)
currency['date'] = pd.to_datetime(currency['date'], format="%d-%m-%Y %H:%M")
currency.set_index('date', inplace=True)

# Resample to monthly frequency and take the mean
currency = currency.resample('M').mean()

# Group by year and check the number of months in each year
groups = currency.groupby(pd.Grouper(freq='Y'))

# Initialize an empty list to store years with fewer than 12 months
incomplete_years = []

for name, group in groups:
    if len(group) < 12:  # Check if there are fewer than 12 months
        incomplete_years.append((name.year, len(group)))

# Display the years with fewer than 12 months and the number of months they contain
print("Years with fewer than 12 months:")
for year, month_count in incomplete_years:
    print(f"Year: {year}, Months: {month_count}")
```

```python
currency = pd.read_csv('exchange_rate.csv', parse_dates=True)
currency['date'] = pd.to_datetime(currency['date'], format="%d-%m-%Y %H:%M")
currency.set_index('date', inplace=True)
down_sample= currency.resample('M').mean()
decomposition = seasonal_decompose(down_sample.Ex_rate, model='additive', period=12)# period 12 means anual seasonality

# Plot decomposition
decomposition.plot()
plt.show()
```

```python
plot_acf(down_sample.Ex_rate,lags=20)
plot_pacf(down_sample.Ex_rate,lags=12)
plt.show()
```

# Data driven Models

```python
len(currency)
```

```python
# train test split
train = currency.head(7000)
test = currency.tail(588)
```

```python
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

# Simple Exponential Method

```python
train["Ex_rate"] = train["Ex_rate"].astype('double')
test["Ex_rate"] = test["Ex_rate"].astype('double') 
ses_model = SimpleExpSmoothing(train["Ex_rate"]).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = test.index[0],end = test.index[-1])
print(MAPE(pred_ses,test.Ex_rate,))
print(mean_absolute_percentage_error(test.Ex_rate,pred_ses)*100)
```

# Holt method

```python
# Holt method 
hw_model = Holt(train["Ex_rate"]).fit(smoothing_level=0.2, smoothing_slope=0.1)
pred_hw = hw_model.predict(start = test.index[0],end = test.index[-1])
MAPE(pred_hw,test.Ex_rate) 
```

# Holts winter exponential smoothing with multiplicative seasonality and multiplicative trend

```python
hwe_model_add_add = ExponentialSmoothing(train["Ex_rate"],seasonal="add",trend="add",seasonal_periods=365).fit() #add the trend to the model
pred_hwe_mul_mul = hwe_model_add_add.predict(start = test.index[0],end = test.index[-1])
MAPE(pred_hwe_add_add,test["Ex_rate"])  
```

# final model in Data driven models

```python
currency["Ex_rate"] = currency["Ex_rate"].astype('double')
hwe_model_mul_mul = ExponentialSmoothing(currency["Ex_rate"],seasonal="add",trend="add",seasonal_periods=365).fit() 
```

```python
original = hwe_model_mul_mul.predict(start=1,end=len(currency))
forcast = hwe_model_mul_mul.forecast(365*10)
```

```python
original.plot(figsize = (24, 5), legend = True)
forcast.plot(legend=True)
```
